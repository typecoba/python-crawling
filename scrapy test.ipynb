```python
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy import Request
```

```python
class TestSpider(scrapy.Spider):
    name = 'test'
    start_urls = [
        'https://smartstore.naver.com/lingerwater/products/2800118009',
    ]

    def start_requests(self):
        headers = {'referer':'https://smartstore.naver.com/lingerwater/products/2800118009'}
        cookies = {}
        yield Request('https://smartstore.naver.com/lingerwater/products/2796364821/reviews/page.json?page=2&size=30&sortType=REVIEW_RANKING&contentType=ALL&topicCode=',
                      headers = headers,
                      cookies = cookies,
                      callback=self.parse, method='GET')

    def parse(self, response):
        print(response.text)
        with open('data_content.json', 'w') as file:
            file.write(response.text)
```

```python
process = CrawlerProcess()
process.crawl(TestSpider)
```

```python
process.start()
```
